#### Tidiness Issue #1: Age stages should be on the same column

#### Define: Allocating a single column that contains age stage for every dog

#### Code

```python
def age_stage(row):
    if row.doggo:
        return 'doggo'
    elif row.floofer:
        return 'floofer'
    elif row.pupper:
        return 'pupper'
    elif row.puppo:
        return 'puppo'
    else:
        return np.NaN
```

#### Test

```python
data_final['dog_stage']=data_final.apply(age_stage,axis=1)
```

#### Tidiness Issue #2:Dog breeds should be on the same column

#### Define: Extracting Dog breeds and storing on a single column

#### Code

```python
def breed(row):
    conf=0
    name=''
    breed=False
    if (row.p1_dog==True):
        if (row.p1_conf>=conf):
            conf=row.p1_conf
            name=row.p1
            breed=True
    if (row.p2_dog==True):
        if (row.p2_conf>=conf):
            conf=row.p2_conf
            name=row.p2
            breed=True
    if (row.p3_dog==True):
        if (row.p3_conf>=conf):
            conf=row.p3_conf
            name=row.p3
            breed=True
    if breed==False:
        return np.NaN
    else:
        return name
```

#### Test

```python
data_final['dog_breed']=data_final.apply(breed,axis=1)
```
